```python
#One-hot encoding to convert all object values to integers & facilitate scaling

def onehot_encode(df,column,prefix):
  data=df.copy()
  dummies=pd.get_dummies(df[column],prefix=prefix)
  data=pd.concat([df,dummies],axis=1)
  data=data.drop(column,axis=1)

  return data
```

```python
for column,prefix in [('destination', 'dest'), ('name', 'name'), ('product_id', 'pid'), ('source', 'src')]:
  cp2 = onehot_encode(cp2,column=column,prefix=prefix)
```

```python
cp2
```

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
lr=LinearRegression()

from sklearn.preprocessing import StandardScaler
scl=StandardScaler()
```

```python
y=cp2['price']
x=cp2.drop('price',axis=1)

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,shuffle=True,random_state=1)
```

```python
x_train
```

```python
# Scaling of training set (value ranges are spread across)
# Standard scaler scales the values to an extent where mean is almost = 0 & variance = 1 for all the records 

scl.fit(x_train)

x_train = pd.DataFrame(scl.transform(x_train),columns=x.columns)
x_test = pd.DataFrame(scl.transform(x_test),columns=x.columns)
```
